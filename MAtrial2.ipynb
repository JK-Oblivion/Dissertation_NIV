## Data loading

### Subtask:
Load the mutation data from the provided CSV file.


**Reasoning**:
I need to load the provided CSV file into a pandas DataFrame.



```python
import pandas as pd

try:
    df = pd.read_csv('ev30_mutation_results.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'ev30_mutation_results.csv' not found.")
    df = None  # Set df to None to indicate failure
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file. Please check the file format.")
    df = None
```

## Data exploration

### Subtask:
Explore the basic statistics and characteristics of the loaded mutation data.


**Reasoning**:
Explore the basic statistics and characteristics of the mutation data by checking dimensions, data types, descriptive statistics, missing values, and unique value counts for relevant columns.



```python
# Check dimensions
print("DataFrame dimensions:", df.shape)

# Examine data types
print("\nData types of each column:\n", df.dtypes)

# Generate descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:\n", df.describe())

# Identify and count missing values
print("\nMissing values per column:\n", df.isnull().sum())

# Analyze 'sequence_name' column
unique_sequence_names = df['Sequence Name'].nunique()
print(f"\nNumber of unique sequence names: {unique_sequence_names}")
sequence_name_counts = df['Sequence Name'].value_counts()
print(f"\nFrequency of each sequence name:\n{sequence_name_counts}")


# Analyze 'mutation_type' column
unique_mutation_types = df['Change Type'].nunique()
print(f"\nNumber of unique mutation types: {unique_mutation_types}")
mutation_type_counts = df['Change Type'].value_counts()
print(f"\nFrequency of each mutation type:\n{mutation_type_counts}")
```

## Data analysis

### Subtask:
Analyze the distribution of mutation types across different sequences.


**Reasoning**:
Group the data by 'Sequence Name' and 'Change Type', count occurrences, calculate relative frequencies, and create a new DataFrame for easier analysis and visualization.



```python
# Group data and count occurrences
mutation_counts = df.groupby(['Sequence Name', 'Change Type']).size().unstack(fill_value=0)

# Calculate total mutations per sequence
mutation_counts['Total'] = mutation_counts.sum(axis=1)

# Calculate relative frequencies
for col in ['Synonymous', 'Non-synonymous']:
    mutation_counts[col + ' (%)'] = (mutation_counts[col] / mutation_counts['Total']) * 100

# Display the resulting DataFrame
display(mutation_counts)

# Sort by non-synonymous mutation frequency
mutation_counts_sorted = mutation_counts.sort_values(by='Non-synonymous (%)', ascending=False)
display(mutation_counts_sorted)
```
